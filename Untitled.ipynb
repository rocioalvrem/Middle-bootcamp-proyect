```python
import pandas as pd

expect = pd.read_excel('expect_compare.xlsx')
expect.head(4)
```

```python
market = pd.read_excel('market_compare.xlsx')
market.head(4)
```

```python
# Aplicar groupby y calcular la media de precios por distrito
mean_prices = market.groupby(['districts','terrace','elevator'])['price'].mean().reset_index()

# Redondear la columna 'price' a dos decimales
mean_prices['price'] = mean_prices['price'].round(2)

# Imprimir el resultado
mean_prices

# Imprimir el result
mean_prices_sorted = mean_prices.sort_values(by='price', ascending=False)

mean_prices_sorted[(mean_prices_sorted['districts']!='canillejas')&(mean_prices_sorted['districts']!='tetuan')&(mean_prices_sorted['districts']!='ciudad_lineal')]

mean_prices_sorted.to_excel('mean_market_prices.xlsx')
```
