```python
import os
import pandas as pd

```

```python
distritos = ['canillejas','vallecas','barajas','moratalaz','tetuan','hortaleza','fuencarral','vicalvaro','usera','villaverde','arganzuela','barrio_salamanca','centro','chamberi','moncloa_aravaca','retiro','chamartin','ciudad_lineal','carabanchel','arguelles','latina']

distritos

carpeta = os.getcwd()
```

```python
# DataFrame para almacenar los datos
df_total = pd.DataFrame()

# Iterar sobre los distritos
for distrito in distritos:
    # Construir el nombre del archivo Excel
    nombre_archivo = f"idealista_{distrito}.xlsx"
    ruta_archivo = os.path.join(carpeta, nombre_archivo)

    # Verificar si el archivo existe antes de intentar cargarlo
    if os.path.exists(ruta_archivo):
        # Leer el archivo Excel
        df_distrito = pd.read_excel(ruta_archivo)
        
        # Agregar una columna 'Distrito' con el nombre del distrito
        df_distrito['Distrito'] = distrito

        # Concatenar el DataFrame del distrito al DataFrame total
        df_total = pd.concat([df_total, df_distrito], ignore_index=True)
    else:
        print(f"Archivo no encontrado para el distrito: {distrito}")

# Mostrar el DataFrame total
print(df_total)
```

```python
df_total.dropna(subset=['Price'], inplace=True)
df_total
```

```python
df_total['swimming_pool'] = df_total['Description'].str.lower().str.contains('piscina')
df_total['swimming_pool'] = df_total['swimming_pool'].astype(int)
df_total['porter'] = df_total['Description'].str.lower().str.contains('portero')
df_total['porter'] = df_total['porter'].astype(int)
df_total['terrace'] = df_total['Description'].str.lower().str.contains('terraza')
df_total['terrace'] = df_total['terrace'].astype(int)
df_total['public_transport'] = df_total['Description'].str.lower().str.contains('publico')
df_total['public_transport'] = df_total['public_transport'].astype(int)
df_total['air_conditioner'] = df_total['Description'].str.lower().str.contains('aire acondicionado')
df_total['air_conditioner'] = df_total['air_conditioner'].astype(int)
df_total['exterior'] = df_total['Description'].str.lower().str.contains('exterior')
df_total['exterior'] = df_total['exterior'].astype(int)
df_total['wardrobes'] = df_total['Description'].str.lower().str.contains('armario')
df_total['wardrobes'] = df_total['wardrobes'].astype(int)
df_total['garden'] = df_total['Description'].str.lower().str.contains('jardín')
df_total['garden'] = df_total['garden'].astype(int)
df_total['storage_room'] = df_total['Description'].str.lower().str.contains('trastero')
df_total['storage_room'] = df_total['storage_room'].astype(int)
```
